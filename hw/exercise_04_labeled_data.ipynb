### Labeled data in two dimensions

In this part, you are going to generate labeled datasets with different characteristics by tuning the parameters in the scikit-learn **make_classification** function.


In Cell 02, create a labeled two dimensional dataset using `make_classification`.  Use the following parameters to call `make_classification` and visualize the resulting dataset using Matplotlib's [scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html) function.

```python
# Code Cell 02
n_samples = 1000
n_features=2
n_informative=2
n_redundant=0
n_repeated=0
n_classes=2
n_clusters_per_class=1
class_sep=1
flip_y=0
random_state=42
# below call make_classification with these parameters and create
# a scatter plot of the data

X, y = make_classification(n_samples=n_samples, n_features=n_features, n_informative=n_informative, n_redundant=n_redundant, n_repeated=n_repeated, n_classes=n_classes, n_clusters_per_class=n_clusters_per_class, class_sep=class_sep, flip_y=flip_y, random_state=random_state)

x_feature_1 = X[:, 0]
x_feature_2 = X[:, 1]

plt.scatter(x_feature_1, x_feature_2, c=y, cmap='viridis')

plt.show()
```

### Class separation

The parameter `class_sep` controls the separation of the classes, and hence changes the difficulty of the classification task. The data points can be easily separable when having a higher class separation.

In Cell 03, create a labled two dimensional dataset by using the function make_classification. Please keep all the key variable unchanged except **class_sep**. Set it to a value that will guarantee accurate classification.
Plot the data by using a **scatter** plot.

```python
# Code Cell 03
n_samples = 1000
n_features=2
n_informative=2
n_redundant=0
n_repeated=0
n_classes=2
n_clusters_per_class=1
class_sep=0.1
flip_y=0
random_state=42

# below write code that demonstrates that using a higher class_sep
# leads to an easier classification problem.

X, y = make_classification(n_samples=n_samples, n_features=n_features, n_informative=n_informative, n_redundant=n_redundant, n_repeated=n_repeated, n_classes=n_classes, n_clusters_per_class=n_clusters_per_class, class_sep=10, flip_y=flip_y, random_state=random_state)

x_feature_1 = X[:, 0]
x_feature_2 = X[:, 1]

plt.scatter(x_feature_1, x_feature_2, c=y, cmap='viridis')

plt.show()
```

In the next code cell we look at the effect of the `weights` parameter of `make_classification`.
In Cell 04, use the following parameter values to set up your dataset and visualize it using a scatter plot.  How does this parameter affect the resulting dataset?  What is the result of using the default value like we did above?

*It looks like it changes the distribution of the data, one class seems to have 90% (the value .9 from weights[0]) of the points and the other class has 10% of the poinnts.  So using the default probably does a 50.50 split for the point distribution in classes.

```python
# Code Cell 04
n_samples = 1000
n_features=2
n_informative=2
n_redundant=0
n_repeated=0
n_classes=2
n_clusters_per_class=1
class_sep=2
flip_y=0
weights=[0.9,0.1] 
random_state=42

X, y = make_classification(n_samples=n_samples, n_features=n_features, n_informative=n_informative, n_redundant=n_redundant, n_repeated=n_repeated, n_classes=n_classes, n_clusters_per_class=n_clusters_per_class, class_sep=class_sep, flip_y=flip_y, weights=weights, random_state=random_state)

x_feature_1 = X[:, 0]
x_feature_2 = X[:, 1]

plt.scatter(x_feature_1, x_feature_2, c=y, cmap='viridis')

plt.show()
```

In the next cell, the focus is on the parameter **flip_y**.  Use the following parameters to set up your dataset and plot the data by using a scatter plot.  What kind of real-world situation does this parameter simulate?

*It looks like noise is added to the data when you change the flip_y value to be closer to 1.  This probably mimicks either errors in dataset creation or/and things like false positives, false negatives.  For exmaple a very poor covid test would have a high flip_y like scenario.*

```python
# Code Cell 05
n_samples = 1000
n_features=2
n_informative=2
n_redundant=0
n_repeated=0
n_classes=2
n_clusters_per_class=1
class_sep=2
flip_y=0.2
random_state=42

X, y = make_classification(n_samples=n_samples, n_features=n_features, n_informative=n_informative, n_redundant=n_redundant, n_repeated=n_repeated, n_classes=n_classes, n_clusters_per_class=n_clusters_per_class, class_sep=class_sep, flip_y=.91, random_state=random_state)

x_feature_1 = X[:, 0]
x_feature_2 = X[:, 1]

plt.scatter(x_feature_1, x_feature_2, c=y, cmap='viridis')

plt.show()

```
